### reshaping

```python
df = df.drop(columns=['Kreiskennziffer', 'Kreis', 'Bundesland'])
df.head()
```

```python
df[['N-Gebiet (=1)', 'Fördergebiet (Ja=1; Nein=0)']][:20]
```

ok, so "N-Gebiet" means not eligible for funding support.

```python
df[['Foerdergebietsstatus', 'Fördergebiet (Ja=1; Nein=0)']][:10]
```

but if you are between "C/N" or "D/N" you are still count as eligible.

so there are three levels, "C", "D", and "N". The category of each kreis can be a single letter or a combination of two letters. Either way, only the status of pure "N" is coded as not eligible for support (Fördergebiet, nein=0).

```python
df = df[['Regionalschlüssel', 'Foerdergebietsstatus', 'Fördergebiet (Ja=1; Nein=0)']]
df.head(3)
```

```python
df = df.rename(columns={'Regionalschlüssel': 'ags5', 
                       'Foerdergebietsstatus': 'support_area_status',
                       'Fördergebiet (Ja=1; Nein=0)': 'eligible_area'})
df.head(3)
```

```python
df.to_csv('../final_dfs/funding_eligibility.csv')
```
